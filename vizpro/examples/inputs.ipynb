# Inputs
Widgets auxiliares para poder interactuar con el gráfico de manera sencilla

```python
from vizpro import RadioInput, SelectInput, CheckboxInput, RangeInput

# Radio buttons (equivalente a tu ejemplo)
metric = RadioInput(
    options=[("Absolute", "absolute"), ("Relative", "relative")], #Lista de tuplas (etiqueta, valor)
    value="absolute",#Opcional, si no se especifica, se selecciona el primero
    label="Escoge una métrica:" # Opcional
)
metric
```

```python
metric.value # Acceder al valor seleccionado
```

```python
range_slider = RangeInput(
    min_val=0,
    max_val=100,
    step=1,
    value=0, #Opcional, si no se especifica, se selecciona el mínimo
    label="Range" # Opcional
)
range_slider
```

```python
range_slider.value  # Acceder al rango seleccionado
```

```python
select_input = SelectInput(
    options=[("Option 1", "opt1"), ("Option 2", "opt2"), ("Option 3", "opt3")],
    value="opt1", #Opcional, si no se especifica, se selecciona
    label="Select an option: " # Opcional
)
select_input
```

```python
select_input.value
```

```python
checkbox_input = CheckboxInput(
    options=["red","green","blue"], #Tambien puede ser una lista simple
    value=["red"], #Opcional, si no se especifica, no se selecciona nada
    label="Check me!", # Etiqueta del checkbox
)
checkbox_input
```

```python
checkbox_input.value
```
