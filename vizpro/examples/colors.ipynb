# Widget para seleccionar una Paleta de colores

Basado en los inputs de esquemas de colores creados en D3.js por Mike Bostock en [Observable](https://observablehq.com/@d3/color-schemes). Incluye todas las escalas categóricas de Color Brewer.

```python
from vizpro import SwatchesInput, CustomWidget
import vizpro.datasets as ds
import traitlets
color_picker1 = SwatchesInput()

# Visualizar
color_picker1
```

```python
color_picker1.palette_name = "Dark2"
```

Ejemplo de Custom Widget extraido del notebook [custom](custom.ipynb)

```python
df_explorers = ds.read_dataset(r"https://gist.githubusercontent.com/MatiasMaravi/5e540b21eb04ca980d77199317e56ebb/raw/Explorers.tsv",sep="\t")
class CustomBarplot_2(CustomWidget):
    _esm = CustomWidget.createWidgetFromLocalFile(paramList=["data","x_","y_","pallete"], filePath=r"..\samples\d3_barchart_modified.js")
    
    data = traitlets.List([]).tag(sync=True)
    x_ = traitlets.Unicode("").tag(sync=True)
    y_ = traitlets.Unicode("").tag(sync=True)
    pallete = traitlets.List([]).tag(sync=True)

barplot2 = CustomBarplot_2(data=df_explorers.data, x_="explorer", y_="frequency", pallete=color_picker1.value)
```

```python
barplot2
```

De por si el grafico es reactivo a cambios en los atributos. Si modificamos directamente la paleta de colores, esto se ve reflejado en el gráfico en tiempo real. Simulando así la experiencia en Observable.

```python
barplot2.pallete = ["#ff0000", "#00ff00", "#0000ff", "#ffff00", "#00ffff", "#ff00ff"]
```

```python
barplot2.pallete = ["#000000", "#333333", "#666666", "#999999", "#CCCCCC", "#FFFFFF"]
```

Si queremos que el grafico sea reactivo a nuestro widget "SwatchesInput", es necesario usar la función "link" de libreria "trailets". Esto con el fin de que nuestro widget escuche los cambios de otro widget.

```python
# linkeamos el "value" de nuestro widget con el atributo de nuestro gráfico
traitlets.link((color_picker1, 'value'), (barplot2, 'pallete'))
```

```python
from vizpro import MatrixCreator, MatrixLayout

m = MatrixCreator(4,4) # Las 2 primeras columnas para el swatches y las 2 ultimas columnas para el barplot
m
```

```python
l = MatrixLayout(m.data)
l.add(color_picker1,1)
l.add(barplot2,2)
```
