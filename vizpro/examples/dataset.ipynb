Así como la libreria tiene por defecto algunos datasets de prueba, el usuario también puede importar sus propios datasets especificando una ruta.

```python
# Importar un dataset desde una URL completa
df_navegadores = Datasets.read_dataset(r"https://gist.githubusercontent.com/MatiasMaravi/5e540b21eb04ca980d77199317e56ebb/raw/Explorers.tsv",sep="\t")
df_navegadores.head()
```
