# Inputs
Auxiliar widgets to interact with the graph in a simple way

```python
from D3Bridge import RadioInput, SelectInput, CheckboxInput, RangeInput

# Radio buttons
metric = RadioInput(
    options=[("Absolute", "absolute"), ("Relative", "relative")], #List of tuples (label, value)
    value="absolute",#Optional, if not specified, the first option is selected
    label="Escoge una métrica:" # Optional
)
metric
```

```python
metric.value # Get the current value of the input
```

```python
range_slider = RangeInput(
    min_val=0,
    max_val=100,
    step=1,
    value=0, #Optional, if not specified, the first option is selected
    label="Range" # Optional
)
range_slider
```

```python
range_slider.value  # Get the current value of the range slider
```

```python
select_input = SelectInput(
    options=[("Option 1", "opt1"), ("Option 2", "opt2"), ("Option 3", "opt3")],
    value="opt1", #Optional, if not specified, the first option is selected
    label="Select an option: " # Optional
)
select_input
```

```python
select_input.value
```

```python
checkbox_input = CheckboxInput(
    options=["red","green","blue"], #List of options
    value=["red"], #Optional, if not specified, no option is selected
    label="Check me!", # Optional label for the checkbox
)
checkbox_input
```

```python
checkbox_input.value
```
