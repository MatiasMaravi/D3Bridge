````python
from D3Bridge import CustomWidget
import D3Bridge.datasets as ds
import traitlets
````

````python
affairsData = ds.get_dataset("affairs")
affairsData.columns
````

````python
jsPath = "scatterplot.js"
cssPath = "scatterplot.css"


class CustomScatterplot(CustomWidget):
    #_esm = CustomWidget.createWidgetFromLocalFile(paramList=["data", "x", "y", "hue"], filePath=jsPath)
    #_css = cssPath
    _esm = CustomWidget.createWidgetFromUrl(paramList=["data", "x", "y", "hue"], jsUrl=r"https://raw.githubusercontent.com/MatiasMaravi/D3Bridge/main/D3Bridge/samples/scatterplot.js")
    _css = CustomWidget.readFromWeb(r"https://raw.githubusercontent.com/MatiasMaravi/D3Bridge/main/D3Bridge/samples/scatterplot.css")
    
    data = traitlets.List([]).tag(sync=True)
    x = traitlets.Unicode().tag(sync=True)
    y = traitlets.Unicode().tag(sync=True)
    hue = traitlets.Unicode().tag(sync=True)
    clickedValue = traitlets.Unicode().tag(sync=True)
    selectedValues = traitlets.List([]).tag(sync=True)

    def on_select_values(self, callback):
        self.observe(callback, names=["selectedValues"])

    def on_click_value(self, callback):
        self.observe(callback, names=["clickedValue"])

scatterplot = CustomScatterplot(data = affairsData.data, x = 'age', y = 'rating', hue = "children")
````

# How does it work?

In the original code, we have the following lines of code:
```js
function setValue(text) {
  model.set({ clickedValue: text });
  model.save_changes();
}

function setSelectedValues(values) {
  model.set({ selectedValues: values });
  model.save_changes();
}
```
This is intended to connect directly to the attributes of our Custom class.
```py
class CustomScatterplot(CustomWidget):
    _esm = CustomWidget.createWidgetFromLocalFile(paramList=["data", "x", "y", "hue"], filePath=jsPath)
    _css = cssPath
    
    data = traitlets.List([]).tag(sync=True)
    x = traitlets.Unicode().tag(sync=True)
    y = traitlets.Unicode().tag(sync=True)
    hue = traitlets.Unicode().tag(sync=True)

    # These new attributes allow us to interact with the data and return data from the "javascript" backend to our Python environment.
    clickedValue = traitlets.Unicode().tag(sync=True)
    selectedValues = traitlets.List([]).tag(sync=True)

    # Functions necessary to ensure responsiveness. If the selected value changes, it will be automatically saved in the specified variables.
    def on_select_values(self, callback):
        self.observe(callback, names=["selectedValues"])

    def on_click_value(self, callback):
        self.observe(callback, names=["clickedValue"])
```
