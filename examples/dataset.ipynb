# **Datasets**

This module aims to simulate the datasets used in D3.js. To do this, it's important to understand that JavaScript interprets a dataset as a list of **objects**.

To simulate this, we need a list of dictionaries.

```python
import D3Bridge.datasets as ds

ds.list_available()
```

```python
df_alphabet = ds.get_dataset("alphabet") #List
df_alphabet
```

```python
df_alphabet.columns
```

Just as the library has some default test datasets, the user can also import their own datasets by specifying a path.

```python
# Importing a dataset from a URL
df_navegadores = ds.read_dataset(r"https://gist.githubusercontent.com/MatiasMaravi/5e540b21eb04ca980d77199317e56ebb/raw/Explorers.tsv",sep="\t")
df_navegadores.head()
```

```python
# Importing a dataset from a local file
df_local = ds.read_dataset(r"C:\Users\USUARIO\Desktop\VIZPRO-D3\D3Bridge\datasets\sales.csv")
df_local.head()
```

It's important to note that the `get_dataset` and `read_dataset` functions return a custom class called `DataFrame`. This class itself isn't intended to replace how lists work. That's why there's an attribute called `data`. This attribute contains a list of dictionaries that simulate JavaScript objects.

```python
df_local.data # We need to pass this attribute to our custom graphs.
```

```python
df_local.min("value") #Value is the name of the column we want to get the minimum value from.
```

```python
df_local.max("value")#Value is the name of the column we want to get the maximum value from.
```

```python
df_local["value"] #Get all values ​​from the 'value' column as a list
```

```python
df_local[0] # Get the first row of the dataset as a dictionary. The keys of the dictionary are the column names and the values are the corresponding values for that row.
```

```python
df_pandas = df_local._data #Accessing the underlying pandas DataFrame of our dataset.
df_pandas.head() #Show the first rows of the pandas DataFrame
```
