```python
import D3Bridge as ds

iris = ds.get_dataset("iris")
iris
```

```python
from D3Bridge import SwatchesInput
```

```python
color_picker = SwatchesInput()
color_picker
```

```python
from D3Bridge import RadioInput
r = RadioInput(options=["vertical", "horizontal"], value="vertical")
r
```

```python
from D3Bridge import BarPlot

chart = BarPlot(
    data=iris.data,
    x="sepal_width",
    y="sepal_length", 
    direction=r.value,
    palette=color_picker.value  #Use the value of the color picker as the palette for the bar plot
)
chart
```

```python
r
```

```python
import traitlets
traitlets.link((color_picker, "value"), (chart, "palette"))
traitlets.link((r, "value"), (chart, "direction"))
```

```python
import xgboost
import shap
# train XGBoost model
X, y = shap.datasets.adult()
model = xgboost.XGBClassifier(n_estimators=100, max_depth=2).fit(X, y)

# compute SHAP values
explainer = shap.Explainer(model, X)
shap_values = explainer(X[:1000])
```

```python
shap_values.values[0]
```

```python
from D3Bridge import BeeswarmPlot
beeswarm = BeeswarmPlot(explanation=shap_values[:200])
beeswarm
```

```python
beeswarm.selected_values
```

```python
from D3Bridge import DecisionPlot
decisionPlot = DecisionPlot(explanation=shap_values[:20])
decisionPlot
```

```python
decisionPlot.selected_values
```

```python
from D3Bridge import HistogramPlot

chart2 = HistogramPlot(
    data=iris.data,
    x="petal_length",
    color= "steelblue" #color_picker.value[0]
)
chart2
```

```python
from D3Bridge import LinearPlot

chart3 = LinearPlot(
    data=iris.data,
    x="sepal_length",
    y="sepal_width",
    hue="species",
    palette=color_picker.value
)
chart3
```

```python
chart3.selected_values
```

```python

df = ds.get_geo_dataset("Brazil")
```

```python
from D3Bridge import GeoMapPlot
# The configuration is optional, but it allows you to customize the appearance of the map.
mapa = GeoMapPlot(df,config={"labelProperty": "name", #property in the dataset to use as label
                             "projectionType": "mercator", #projection type to use
                             "hoverFillColor": "orange",# color to use when hovering over a region
                             "strokeColor": "black",# color to use for the stroke of the regions
                             "fillColor": "steelblue", #color to use for the fill of the regions
                             "strokeWidth": 1}) #width of the stroke of the regions
mapa
```

```python
mapa.selected_region
```
