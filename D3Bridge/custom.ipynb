```python
from vizpro import CustomWidget, MatrixLayout, MatrixCreator
import traitlets
```

```python

path = "samples/d3_barchar2.js"

class CustomBarplot(CustomWidget):
    _esm = CustomWidget.createWidgetFromLocalFile(paramList=["data", "x", "y", "hue"], filePath=path)
    
    #_esm = CustomWidget.createWidgetFromUrl(paramList=["data", "x", "y", "hue"], jsUrl=url)
    
    data = traitlets.List([]).tag(sync=True)
    x = traitlets.Unicode().tag(sync=True)
    y = traitlets.Unicode().tag(sync=True)
    hue = traitlets.Unicode().tag(sync=True)

barplot = CustomBarplot(data=df.to_dict(orient="records"), x='letter', y='frequency')
```

```python
from vizpro import RadioInput, RangeInput

# Radio buttons (equivalente a tu ejemplo)
metric = RadioInput(
    options=[("Absolute", "absolute"), ("Relative", "relative")],
    value="absolute",
    label="Change"
)

range_slider = RangeInput(
    min_val=2010,
    max_val=2019,
    step=1,
    value=[2010, 2019],
    label="Year Range"
)
```

```python

path = "samples/d3_barchar3.js"

class CustomBarplot3(CustomWidget):
    # Agregar 'metric' a la lista de parámetros
    _esm = CustomWidget.createWidgetFromLocalFile(paramList=["data", "metric"], filePath=path)

    data = traitlets.List([]).tag(sync=True)
    metric = traitlets.Unicode("absolute").tag(sync=True)  # Nuevo parámetro

barplot2 = CustomBarplot3(data=df_2.to_dict(orient="records"), metric="absolute")
```

```python
# Vincular el RadioInput con el gráfico
traitlets.link((metric, "value"), (barplot2, "metric"))
```
