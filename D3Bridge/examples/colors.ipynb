# Widget to select a color palette

Based on the color scheme inputs created in D3.js by Mike Bostock in [Observable](https://observablehq.com/@d3/color-schemes). It includes all the categorical Color Brewer scales.

```python
from D3Bridge import SwatchesInput, CustomWidget
import D3Bridge.datasets as ds
import traitlets
color_picker1 = SwatchesInput()

# Visualize the widget
color_picker1
```

```python
color_picker1.value
```

```python
color_picker1.palette_name = "Dark2"
```

Ejemplo de Custom Widget extraido del notebook [custom](custom.ipynb)

```python
df_explorers = ds.read_dataset(r"https://gist.githubusercontent.com/MatiasMaravi/5e540b21eb04ca980d77199317e56ebb/raw/Explorers.tsv",sep="\t")
class CustomBarplot_2(CustomWidget):
    _esm = CustomWidget.createWidgetFromLocalFile(paramList=["data","x_","y_","palette"], filePath=r"..\samples\d3_barchart_modified.js")
    
    data = traitlets.List([]).tag(sync=True)
    x_ = traitlets.Unicode("").tag(sync=True)
    y_ = traitlets.Unicode("").tag(sync=True)
    palette = traitlets.List([]).tag(sync=True)

barplot2 = CustomBarplot_2(data=df_explorers.data, x_="explorer", y_="frequency", palette=color_picker1.value)
```

```python
barplot2
```

The graph is reactive to changes in the attributes. If we directly modify the color palette, this is reflected in the graph in real time. Simulating the experience in Observable.

```python
barplot2.palette = ["#ff0000", "#00ff00", "#0000ff", "#ffff00", "#00ffff", "#ff00ff"]
```

```python
barplot2.palette = ["#000000", "#333333", "#666666", "#999999", "#CCCCCC", "#FFFFFF"]
```

Si queremos que el grafico sea reactivo a nuestro widget "SwatchesInput", es necesario usar la función "link" de libreria "trailets". Esto con el fin de que nuestro widget escuche los cambios de otro widget.

```python
# linkeamos el "value" de nuestro widget con el atributo de nuestro gráfico
traitlets.link((color_picker1, 'value'), (barplot2, 'palette'))
```

```python
from D3Bridge import MatrixCreator, MatrixLayout

m = MatrixCreator(2,4) # 2 first columns for the swatches and 2 last columns for the barplot
m
```

```python
l = MatrixLayout(m.data)
l.add(color_picker1,1)
l.add(barplot2,2)
```
