```python
import json
path_ = "../datasets/states-albers-10m.json"
mapa_data = json.load(open(path_))
```

```python
path_map = r"..\samples\map.js"

class CustomMapa(CustomWidget):
    _esm = CustomWidget.createWidgetFromLocalFile(paramList=["data"], filePath=path_map,
                                                  extra_imports=['import * as topojson from "https://esm.sh/topojson-client@3";'])
    data = traitlets.Dict(mapa_data).tag(sync=True)
    selectedValues = traitlets.List([]).tag(sync=True)

    def on_select_values(self, callback):
        self.observe(callback, names=["selectedValues"])
    

map_widget = CustomMapa(data = mapa_data)
```

```python
iris = ds.get_dataset("iris")
iris.head()
```

```python
from D3Bridge import ScatterPlot

scatterplot = ScatterPlot(
    data=iris.data,
    x="sepal_length",
    y="sepal_width",
    hue="species",
    palette=['#1f77b4','#ff7f0e','#2ca02c']
)
scatterplot
```

```python
from D3Bridge import BarPlot
barplot1 = BarPlot(
    data=iris.data,
    x="sepal_length",
    y="sepal_width",
    hue="species",
    palette=['#1f77b4','#ff7f0e','#2ca02c']
)
barplot1
```

```python
from D3Bridge import HistogramPlot

histplot = HistogramPlot(
    data=iris.data,
    x="petal_length",
    color= "steelblue" #color_picker.value[0]
)
histplot
```

```python
from D3Bridge import LinearPlot

lineplot = LinearPlot(
    data=iris.data,
    x="sepal_length",
    y="sepal_width",
    hue="species",
    palette=['#1f77b4','#ff7f0e','#2ca02c']
)
lineplot
```

```python
from D3Bridge import SwatchesInput
color_picker = SwatchesInput()
color_picker
```

```python
from D3Bridge import MatrixLayout

layout = MatrixLayout(matrix.data)

layout.add(scatterplot,1)
layout.add(barplot1,2)
layout.add(histplot,3)
layout.add(lineplot,4)

layout
```
