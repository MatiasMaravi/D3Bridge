```python
import vizpro.datasets as ds

iris = ds.get_dataset("iris")
iris
```

```python
from vizpro import SwatchesInput
```

```python
color_picker = SwatchesInput()
color_picker
```

```python
from vizpro import RadioInput
r = RadioInput(options=["vertical", "horizontal"], value="vertical")
r
```

```python
from vizpro.graphs import BarPlot

chart = BarPlot(
    data=iris.data,
    x="sepal_width",
    y="sepal_length", 
    direction=r.value,  # o "horizontal"
    palette=color_picker.value  # Usar la paleta seleccionada en el color picker
)
chart
```

```python
r
```

```python
import traitlets
traitlets.link((color_picker, "value"), (chart, "palette"))
traitlets.link((r, "value"), (chart, "direction"))
```

```python
import xgboost
import shap
# train XGBoost model
X, y = shap.datasets.adult()
model = xgboost.XGBClassifier(n_estimators=100, max_depth=2).fit(X, y)

# compute SHAP values
explainer = shap.Explainer(model, X)
shap_values = explainer(X[:1000])
```

```python
shap_values.values[0]
```

```python
from vizpro import BeeswarmPlot
beeswarm = BeeswarmPlot(explanation=shap_values[:200])
beeswarm
```

```python
beeswarm.selected_values
```

```python
from vizpro import DecisionPlot
decisionPlot = DecisionPlot(explanation=shap_values[:20])
decisionPlot
```

```python
decisionPlot.selected_values
```
